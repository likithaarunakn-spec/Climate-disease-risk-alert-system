```python
plt.figure(figsize=(6,4))
sns.histplot(df["malaria_cases"], bins=30,kde=True)
plt.title("Distribution of Malaria Cases")
plt.xlabel("Malaria Cases")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["dengue_cases"], bins=30, kde=True)
plt.title("Distribution of Dengue Cases")
plt.xlabel("Dengue Cases")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=df["avg_temp_c"],
    y=df["malaria_cases"],
    alpha=0.5
)
plt.title("Average Temperature vs Malaria Cases")
plt.xlabel("Average Temperature (C)")
plt.ylabel("Malaria Cases")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=df["precipitation_mm"],
    y=df["dengue_cases"],
    alpha=0.5
)
plt.title("Rainfall vs Dengue Cases")
plt.xlabel("Precipitation (mm)")
plt.ylabel("Dengue Cases")
plt.show()
```

```python
monthly_malaria = df.groupby("month")["malaria_cases"].mean()
plt.figure(figsize=(6,4))
monthly_malaria.plot(kind="line", marker="o")
plt.title("Average Monthly Malaria Cases")
plt.xlabel("Month")
plt.ylabel("Average Malaria Cases")
plt.show()
```

```python
monthly_dengue = df.groupby("month")["dengue_cases"].mean()
plt.figure(figsize=(6,4))
monthly_dengue.plot(kind="line", marker="o", color="orange")
plt.title("Average Monthly Dengue Cases")
plt.xlabel("Month")
plt.ylabel("Average Dengue Cases")
plt.show()
```

```python
top_malaria = (
    df.groupby("country")["malaria_cases"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
plt.figure(figsize=(8,4))
top_malaria.plot(kind="bar")
plt.title("Top 10 Countries by Total Malaria Cases")
plt.ylabel("Total Cases")
plt.show()
```

```python
top_dengue = (
    df.groupby("country")["dengue_cases"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
plt.figure(figsize=(8,4))
top_dengue.plot(kind="bar", color="red")
plt.title("Top 10 Countries by Total Dengue Cases")
plt.ylabel("Total Cases")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(
    df[
        [
            "avg_temp_c",
            "precipitation_mm",
            "air_quality_index",
            "uv_index",
            "population_density",
            "malaria_cases",
            "dengue_cases"
        ]
    ].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Between Climate Factors and Disease Cases")
plt.show()
```
